```python
from bs4 import BeautifulSoup as bs
import requests
from pdfminer.high_level import extract_text
import json
```

```python
root = 'https://www.icc-cpi.int'
d = "/case-transcripts?f%5B0%5D=court_phase_of_case%3A131&f%5B1%5D=t_case_code%3A1147"
```

```python
url = root + d
r = requests.get(url)
```

```python
soup = bs(r.text)
```

```python
subsites = [root + x['href'] for x in soup.findAll("a") if x.get("href") and "court-record" in x.get("href") and "en" in x.get("href")]
```

```python
subsoups = [bs(requests.get(x).text) for x in subsites]
```

```python
pdfs = [[x['href'] for x in y.findAll("a") if x.get("href") and "Transcripts" in x.get("href")][0] for y in subsoups]
```

```python
names = [i[-14:-4] for i in pdfs]
```

```python
for (f,n) in zip(pdfs, names):
    r = requests.get(root + f)
    with open(n + ".pdf", 'wb') as f:
        f.write(r.content)
```

```python
docs = {n: extract_text(n + ".pdf") for n in names}
```

```python
with open("text.json", 'w+') as file:
    json.dump(docs, file)
```
